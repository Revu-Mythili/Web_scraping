```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
```

```python
page = requests.get(url)
```

```python
print(page)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('title')
```

```python
soup.find('div', class_ = 'mw-heading mw-heading2').text
```

```python
soup.find('table')
```

```python
# find - first table, find_all - all tables in list 
# if we have multiple tables to get information from 2nd or 3rd table - we can use find_all - gives us list - specify position start at 0,1,2...
soup.find_all('table')
```

```python
soup.find_all('table')[0]
```

```python
# or we can use class if tables have different classes
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
# th - column names
column_names = table.find_all('th')
```

```python
column_names
```

```python
# list comprehension
table_column_titles = [title.text.strip() for title in column_names]
```

```python
table_column_titles
```

```python
#loop 
for row in all_data:
    rows = row.find_all('td')
    table_row_data = [data.text.strip() for data in rows] 
print(table_row_data)
```

```python
for row in all_data[1:]:
    rows = row.find_all('td')
    table_row_data = [data.text.strip() for data in rows] 
    print(table_row_data)
```

```python
# looping through tr data (all_data) - looking for td tags - individual data (rows) - getting data text and strip it 
# (now its in list for each individual row) - looking for length dataframe - appending each row information into next position 
# len - length loc - location (index of dataframe - how many rows are there in data frame)
for row in all_data[1:]:
    rows = row.find_all('td')
    table_row_data = [data.text.strip() for data in rows] 
    
    length = len(df)
    print(length)
    df.loc[length] = table_row_data
```
